```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("loan_approval_data.csv")
```

```python
df.info()
```

```python
num_col = df.select_dtypes(include=["float"]).columns
obj_col = df.select_dtypes(include=["object"]).columns
```

```python
obj_col
```

```python
from sklearn.impute import SimpleImputer
num_cols = SimpleImputer(strategy="mean")
df[num_col] = num_cols.fit_transform(df[num_col])
obj_cols = SimpleImputer(strategy="most_frequent")
df[odj_col] = obj_cols.fit_transform(df[obj_col])
```

```python
df.isnull().sum()
```

```python
classes_count = df["Loan_Approved"].value_counts()
plt.pie(classes_count, labels=["No", "Yes"], autopct="%1.1f%%")
plt.title("Is Loan approved or not?")
 
 
```

```python
edu_cnt = df["Education_Level"].value_counts()
ax = sns.barplot(edu_cnt)
ax.bar_label(ax.containers[0])
```

```python
Employer_Cat =df["Employment_Status"].value_counts()
ax = sns.barplot(Employer_Cat)
ax.bar_label(ax.containers[0])
```

```python
Employer_Cate =df["Employer_Category"].value_counts()
ax = sns.barplot(Employer_Cate)
ax.bar_label(ax.containers[0])
```

```python
ln =df["Loan_Term"].value_counts()
ax = sns.barplot(ln)
ax.bar_label(ax.containers[0])
```

```python
fig, axes = plt.subplots(2, 2)

sns.boxplot(ax=axes[0, 0], data=df, x="Loan_Approved",y="Applicant_Income")
sns.boxplot(ax=axes[0, 1], data=df, x="Loan_Approved",y="Credit_Score")
sns.boxplot(ax=axes[1, 0], data=df, x="Loan_Approved",y="DTI_Ratio")
sns.boxplot(ax=axes[1, 1], data=df, x="Loan_Approved",y="Savings")

plt.tight_layout()
```
